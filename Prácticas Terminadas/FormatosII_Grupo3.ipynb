Para poder utilizar la API de servicios, en primre lugar hay que registrarse, lo cual puede hacerse en la página http://opendata.emtmadrid.es/Formulario. El registro devuelve como resultado en un mensaje electrónico, dos valores:

* Id de cliente: Identificador de cliente

* Pass Key: Password

A continuación, hay que elegir el servicio que se quiere utilizar. Hay 4 servicios definidos: BUS, GEO, MEDIA, INFOPARKING y BICIMAD. Cada servicio tiene asociado un conjunto de métodos que al invocarlos, devuelven un resultado.
```python
import requests

# Importamos las librerías necesarias para trabajar con el modelo DOM:
import xml.dom.minidom
from xml.dom.minidom import parse, parseString

# Funciones proporcionadas en el enunciado:
def getStreet(calle):
    data = {
        'idClient':'WEB.SERV.mapast03@ucm.es',
        'PassKey':'B60FB34A-DBF8-4463-95EF-4010BC3D43AC',
        'description': calle,
        'streetNumber':'',
        'Radius':'',
        'Stops':'',
        'statistics':'',
        'cultureInfo':''
    }

    url = 'https://servicios.emtmadrid.es:8443/geo/servicegeo.asmx/GetStreet'
    return requests.post(url, data=data)
    # La funcion getStreet devuelve un documento XML, entre cuyos elementos se
    # encuentran las coordenadas X e Y del lugar buscado.

def getStreetRoute(origenX,origenY,destinoX,destinoY):
    data = {
        'idClient':'WEB.SERV.mapast03@ucm.es',
        'PassKey':'B60FB34A-DBF8-4463-95EF-4010BC3D43AC',
        'statistics':'',
        'cultureInfo':'',
        'coordinateXFrom': origenX,
        'coordinateYFrom': origenY,
        'originName':'',
        'coordinateXTo': destinoX,
        'coordinateYTo': destinoY,
        'destinationName':'',
        'criteriaSelection':'11',
        'day':'',
        'month':'',
        'year':'',
        'hour':'',
        'minute':'',
        'GenerarAudio':''
    }

    url = 'https://servicios.emtmadrid.es:8443/servicemedia/servicemedia.asmx/GetStreetRoute'
    return requests.post(url, data=data)

# [1 punto]
# Crear una función que solicite al usuario dos calles de Madrid, que actúen de
# origen y destino de una ruta:
def solicitaCalles():
    origen = input("Introduzca calle de origen: ")
    destino = input("Introduzca calle de destino: ")
    return origen, destino

# [2 puntos]
# Crear una función que dada una calle de Madrid, devuelva las coordenadas X e Y
# de la calle:
def dameCoord(calle):
    response = getStreet(calle)
    docXML = parseString(response.text)
    street = docXML.documentElement

    localiz = street.getElementsByTagName("Site")[0]

    coorX = localiz.getElementsByTagName("CoordinateX")[0]
    X = coorX.childNodes[0].data

    coorY = localiz.getElementsByTagName("CoordinateY")[0]
    Y = coorY.childNodes[0].data

    return X, Y

# [4 puntos]
# Crear una función que dadas las coordenadas X e Y de un origen y un destino
# dentro de Madrid, obtenga la ruta más óptima para ir del origen al destino
# usando autobuses de la EMT:
def getRouteFromCoords(origenX, origenY, destinoX, destinoY):
    # Abrimos el documento XML que devuelve la API de la EMT con el método parse,
    # que nos proporciona un árbol DOM del documento.
    response = getStreetRoute(origenX,origenY,destinoX,destinoY)
    docXML = parseString(response.text)

    # Accedemos a la raiz del árbol y a sus diferentes nodos hojas:
    street = docXML.documentElement

    # RouteData
    ruta = street.getElementsByTagName("ListRouteData")[0]

    datosRuta = ruta.getElementsByTagName("DescriptionRouteData")
    # Para acceder al contenido de cada elemento, utilizamos el atributo data del objeto childNodes
    print (datosRuta[0].getElementsByTagName("DescriptionDate")[0].childNodes[0].data)
    print (datosRuta[0].getElementsByTagName("DescriptionInitTime")[0].childNodes[0].data)
    print ("Hora estimada de llegada: ",datosRuta[0].getElementsByTagName("DescriptionEstimateTimeArrival")[0].childNodes[0].data, "\n")
    print ("Número de trasbordos: ",datosRuta[0].getElementsByTagName("Transfers")[0].childNodes[0].data)
    print ("Duración del viaje: ",datosRuta[0].getElementsByTagName("LongJourney")[0].childNodes[0].data, "\n")
    print ("Descripción textual de la ruta: ")

    # Section
    rutaSecList = ruta.getElementsByTagName("ListSectionRoute")[0]
    section = rutaSecList.getElementsByTagName("Section")

    j = 0
    for i in section:
        if section[j].getElementsByTagName("WalkingLeg"):
            # Si la API de la EMT ha devuelto la tag WalkingLeg significa que tenemos
            # que recorrer un tramo a pie, ya sea para llegar a otra parada a la
            # que hacer transbordo o porque se tarda menos a pie que cogiendo un
            # autobus:
            print("    ", section[j].getElementsByTagName("WalkingLeg")[0].getElementsByTagName("RouteDescription")[0].childNodes[0].data)
        else:
            # En caso contrario, devuelve la información del bus que hay que coger
            print("    ", section[j].getElementsByTagName("BusLeg")[0].getElementsByTagName("RouteDescription")[0].childNodes[0].data)

        j+=1

    print("\n");

    return response

# [1 punto]
# Crear una función que solicite al usuario los nombres de 2 calles de Madrid y
# muestre la información referente a la ruta más óptima para llegar de una calle
# a otra usando autobuses de la EMT:
def getRouteFromStreets():
    origen, destino = solicitaCalles()

    origenX, origenY = dameCoord(origen)
    destinoX, destinoY = dameCoord(destino)

    return getRouteFromCoords(origenX, origenY, destinoX, destinoY)

# [2 puntos]
# Crear una función que solicite al usuario los nombres de 2 calles de Madrid y
# muestre un listado de todos los puntos de interés que hay dentro de la ruta más
# óptima para llegar de una calle a otra usando autobuses de la EMT:
def getPointsOfInterest():
    response = getRouteFromStreets()
    docXML = parseString(response.text)
    resultado = docXML.documentElement

    # La función GetStreetRoute de la API de la EMT devuelve un documento XML con
    # mucha información, entre la que se encuentra un subnodo con los PDIs:
    listPOIS = []
    POIS = resultado.getElementsByTagName("POIS")[0].getElementsByTagName("ListPOIS")[0].getElementsByTagName("POI") # ListPOIS
    for poi in POIS:
        listPOIS.append(poi.getElementsByTagName("nombre")[0].childNodes[0].data)

    print ("Estos son los puntos de interés de su trayecto:\n")
    for poi in listPOIS:
        print("    ", poi)

    consulta = input ("\nEscriba el punto de interés del que quiere recibir información: ")
    for poi in POIS:
        if consulta == poi.getElementsByTagName("nombre")[0].childNodes[0].data:
            # Puede ocurrir que haya dos PDIs con el mismo nombre (por ejemplo,
            # establecimientos que sean franquicias como puede ser el GINOS o el
            # RODILLA). En ese caso, se muestra la información para todos aquellos
            # PDIs que coincidas con el nombre introducido por el usuario:
            print("\n    Nombre: ",poi.getElementsByTagName("nombre")[0].childNodes[0].data)
            if poi.getElementsByTagName("direccion") and poi.getElementsByTagName("numeroCalle"):
                print("    Dirección: ",poi.getElementsByTagName("direccion")[0].childNodes[0].data, " ", poi.getElementsByTagName("numeroCalle")[0].childNodes[0].data)
            else:
                print("    Dirección: ",poi.getElementsByTagName("direccion")[0].childNodes[0].data)

            if poi.getElementsByTagName("telefono"):
                print("    Teléfono: ", poi.getElementsByTagName("telefono")[0].childNodes[0].data)

if __name__ == "__main__":
        getRouteFromStreets()
        getPointsOfInterest()
```
